### Bag of words approach


Make the words as vectors. Get all the unique words in the tweets/ corpus. Create one sized vector. Count how many times these words appear.

### Step 1: Data clean up

```python
#Remove stop words. 
from collections import Counter
vocab = Counter()
for tweets in train.SentimentText:
    for words in tweets.split(' '):
        vocab[words] += 1
```

```python
(vocab.most_common(25))
```

```python
from bokeh.models import ColumnDataSource, LabelSet
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook
output_notebook()
```

```python
import math
import numpy as np

def plot_distribution(vocabulary):

    hist, edges = np.histogram(list(map(lambda x:math.log(x[1]),vocabulary.most_common())), density=True, bins=500)

    p = figure(tools="pan,wheel_zoom,reset,save",
               toolbar_location="above",
               title="Word distribution accross all tweets")
    p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], line_color="#555555", )
    show(p)

plot_distribution(vocab)
```
